```python
import requests
from bs4 import BeautifulSoup

# Making a GET request
# r = requests.get('https://www.geeksforgeeks.org/python-programming-language/')
r = requests.get('https://remysharp.com/2021/07/09/how-to-inspect-elements-that-hide-on-mouse-move')


# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')

# s = soup.find('div', class_='entry-content')
s = soup.find('div', class_='post-content e-content')

lines = s.find_all('p')

for line in lines:
    print(line.text)
```
